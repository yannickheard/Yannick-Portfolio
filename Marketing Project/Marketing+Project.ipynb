```python
def combine(row):
    row=eval(row)
    row=sorted(row,key=str.lower)
    row=str(row)
    return row
    
def reduce1(row):
    row=eval(row)
    try:
        row=row[0]
    except:
        KeyError
    row=str(row)
    return row
    
```

```python
def ev(row):
    year_lis=[2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
    a=eval(row)
    a=sorted(a)
    for i in a:
        if i in year_lis:
            a=year_lis.index(i)
            year_lis[a]=1
    for t in year_lis:
        if t!=1:
            b=year_lis.index(t)  
            year_lis[b]=0
    return year_lis
```

```python
def pct_rank_qcut(series, n):
    edges = pd.Series([float(i) / n for i in range(n + 1)])
    f = lambda x: (edges >= x).argmax()
    return series.rank(pct=1).apply(f)
```
