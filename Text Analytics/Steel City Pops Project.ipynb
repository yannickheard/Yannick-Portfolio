### Some pre-processing: removing stop words and lemmatizing

```python
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
lemma = WordNetLemmatizer()

def reviews_to_words(message):
    # Function to convert a raw message to a string of words
    #
    # Remove non-letters 
    letters_only = re.sub("[^a-zA-Z]", " ", str(message)) 
    #
    # Convert to lower case, split into individual words
    words = letters_only.lower().split() 
    #    
    # Remove stop words
    stops = set(stopwords.words("english")) 
    non_stop_words = [w for w in words if w not in stops] 
    #
    # Lemmatize
    lemma = WordNetLemmatizer()
    meaningful_words = []
    for w in non_stop_words:
        lemmaword = lemma.lemmatize(w)
        meaningful_words.append(lemmaword)
    
    # Remove duplicates
    uwords = []
    [uwords.append(x) for x in meaningful_words if x not in uwords]
    #
    # Join the words back into one string separated by space, 
    onestring = ( " ".join( uwords ))
     
    # Append to new list "cleanmessages"
    cleanmessages.append(onestring)
    #
    # return the string
    return onestring


cleanmessages = []

Results_df['MessageWords'] = Results_df['Message'].map(reviews_to_words)
Results_df.head()
```

```python
top_10_brandwords = pd.Series(" ".join(scp_df['Brand Words']).split()).value_counts()[:10]
top_10_brandwords
```

```python
def calculatelift(brand1,brand2):
    N = len(scp_df)
    num_brand1 = top_10_brandwords[brand1]
    num_brand2 = top_10_brandwords[brand2]
    num_both = CountTwoBrands(brand1,brand2)
    lift = float(N*num_both)/(num_brand1*num_brand2)
    return lift

def CountTwoBrands(brand1,brand2):
    count = 0
    for i in range(0,len(scp_df)):
        if (brand1 in scp_df.iloc[i]['Brand Words'] and brand2 in scp_df.iloc[i]['Brand Words']):
            count += 1
    return count

my_list = top_10_brandwords.index.tolist()
liftvals = []
for pair in itertools.product(my_list, repeat=2):
    lift = calculatelift(*pair)
    liftvals.append(lift)
```

```python
brandlift_df = pd.DataFrame(np.array(liftvals).reshape(10,10), index = top_10_brandwords.index.tolist(), columns = top_10_brandwords.index.tolist())
for b in range(0,10):
    brandlift_df.iloc[b][top_10_brandwords.index.tolist()[b]] = 1
brandlift_df
```

### From the lift calculations, some observations:

- "Natural" and "Organic" provided a very high lift value of 8.25.

We see that our hypothesis was correct - most people would be interested in multiple "brand attributes" as opposed to just one. For example, if we were interested in buying organic popsicles, we would also hope that they were fresh, simple, natural, gourmet, and local.
